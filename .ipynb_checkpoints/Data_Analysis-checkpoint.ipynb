```python
import pandas as pd

# Import the CSV file into a DataFrame
all_seasons_data = pd.read_csv("all_seasons_data.csv")
```

```python
import matplotlib.pyplot as plt

# Count the number of correct (1) and incorrect (0) predictions
prediction_counts = all_seasons_data['bookmakers_prediction'].value_counts()

print(prediction_counts)

# Plot the results as a bar chart
plt.figure(figsize=(8, 5))
plt.bar(prediction_counts.index, prediction_counts.values, tick_label=["Incorrect", "Correct"], alpha=0.75)
plt.title("Bookmakers Prediction Accuracy")
plt.xlabel("Prediction Outcome")
plt.ylabel("Count")
plt.show()

# Optional: Pie Chart for a different visualization
plt.figure(figsize=(6, 6))
plt.pie(prediction_counts, labels=["Incorrect", "Correct"], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title("Bookmakers Prediction Accuracy")
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Directory containing the season CSV files
data_directory = "./Data_By_Season"

# Function to determine if the bookmaker predicted correctly
def bookmakers_prediction(row):
    # Find the smallest odd
    smallest_odd = min(row['homeodd'], row['drawodd'], row['awayodd'])

    # Determine the winning odd
    if row['home_odd_win'] == 1:
        winning_odd = row['homeodd']
    elif row['draw_odd_win'] == 1:
        winning_odd = row['drawodd']
    elif row['away_odd_win'] == 1:
        winning_odd = row['awayodd']
    else:
        winning_odd = None  # Handle cases where no winning odd is found

    # Compare smallest odd with winning odd
    return 1 if smallest_odd == winning_odd else 0

# Iterate through all CSV files in the directory
for file_name in os.listdir(data_directory):
    if file_name.endswith(".csv"):  # Check if the file is a CSV
        season_name = file_name.replace(".csv", "")  # Use the file name (without extension) as the season name
        file_path = os.path.join(data_directory, file_name)  # Get the full file path
        
        # Load the CSV file into a DataFrame
        season_data = pd.read_csv(file_path)
        
        # Ensure relevant columns are numeric and handle missing data
        columns_to_check = ["homeodd", "drawodd", "awayodd", "home_odd_win", "draw_odd_win", "away_odd_win"]
        season_data[columns_to_check] = season_data[columns_to_check].apply(pd.to_numeric, errors='coerce')
        season_data.fillna(0, inplace=True)  # Replace NaN values with 0

        # Delete rows with null odds
        season_data = season_data[
            ~(
                (season_data["homeodd"] == 0) &
                (season_data["drawodd"] == 0) &
                (season_data["awayodd"] == 0) &
                (season_data["home_odd_win"] == 0) &
                (season_data["draw_odd_win"] == 0) &
                (season_data["away_odd_win"] == 0)
            )
        ]

        # Apply the bookmakers_prediction function
        season_data['bookmakers_prediction'] = season_data.apply(bookmakers_prediction, axis=1)

        # Save the processed DataFrame back to the file
        season_data.to_csv(file_path, index=False)
        
        # Visualize the data
        prediction_counts = season_data['bookmakers_prediction'].value_counts()

        # Bar Chart
        plt.figure(figsize=(8, 5))
        plt.bar(prediction_counts.index, prediction_counts.values, tick_label=["Incorrect", "Correct"], alpha=0.75)
        plt.title(f"Bookmakers Prediction Accuracy ({season_name})")
        plt.xlabel("Prediction Outcome")
        plt.ylabel("Count")
        plt.show()

        # Pie Chart
        plt.figure(figsize=(6, 6))
        plt.pie(prediction_counts, labels=["Incorrect", "Correct"], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
        plt.title(f"Bookmakers Prediction Accuracy ({season_name})")
        plt.show()
```
