```python
import pandas as pd

# Import the CSV file into a DataFrame
all_seasons_data = pd.read_csv("all_seasons_data.csv")
```

```python
# Create the 'win' column using a lambda function
all_seasons_data['winning_class'] = all_seasons_data.apply(
    lambda row: 1 if row['home_odd_win'] == 1 else 
                2 if row['draw_odd_win'] == 1 else 
                3 if row['away_odd_win'] == 1 else 0,
    axis=1
)

# Display the updated DataFrame
display(all_seasons_data.head(20))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Define features (X) and target (Y)
X = all_seasons_data[['homeodd', 'drawodd', 'awayodd']]
Y = all_seasons_data['winning_class']

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Model 1: Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, Y_train)
logistic_predictions = logistic_model.predict(X_test)

print("Logistic Regression Results:")
print("Accuracy:", accuracy_score(Y_test, logistic_predictions))
print(classification_report(Y_test, logistic_predictions))

# Model 2: Random Forest Classifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, Y_train)
random_forest_predictions = random_forest_model.predict(X_test)

print("Random Forest Classifier Results:")
print("Accuracy:", accuracy_score(Y_test, random_forest_predictions))
print(classification_report(Y_test, random_forest_predictions))
```

```python
# Logistic Regression Predictions
logistic_results = pd.DataFrame({
    'Actual': Y_test,
    'Predicted_Logistic': logistic_predictions
})

print("Logistic Regression Results:")
display(logistic_results.head(10))

# Random Forest Predictions
random_forest_results = pd.DataFrame({
    'Actual': Y_test,
    'Predicted_RandomForest': random_forest_predictions
})

print("\nRandom Forest Classifier Results:")
display(random_forest_results.head(10))

# If you want to combine all in one DataFrame:
combined_results = pd.DataFrame({
    'Actual': Y_test.reset_index(drop=True),
    'Predicted_Logistic': logistic_predictions,
    'Predicted_RandomForest': random_forest_predictions
})

print("\nCombined Results:")
display(combined_results)
```
